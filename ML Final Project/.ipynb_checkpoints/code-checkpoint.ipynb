```python
# od.download("https://www.kaggle.com/datasets/uciml/electric-power-consumption-data-set/data")
import pandas as pd

# Specify the path to your text file
text_file_path = 'Dataset/Household Power Consumption.txt'

# Read the text file into a DataFrame using pandas
data = pd.read_csv(text_file_path,sep=';')

# Save the DataFrame to a CSV file
# data.to_csv('dataset.csv', index=False)

# Display first 5 columns
data.head()
```

<h4 style="
  font-family: Arial, Helvetica, sans-serif;

  -webkit-text-fill-color: transparent;
  -webkit-background-clip: text;
">| Data Cleaning</h4>


```python
# print the shape of the dataset
data.shape
```

```python
# check for missing values
data.isnull().sum()
```

```python
data.dropna(subset=['Sub_metering_3'], inplace=True)
```

```python
# make sure the data types are in correct format
data['Date'] = data['Date'].astype(str)
data['Time'] = data['Time'].astype(str)
```

```python
# replace '?' and 'nan' with -1
data.replace(['?', 'nan', np.nan,'NaN',"   ","  "," "], '-1', inplace=True ) 
```

```python
data.dtypes
```

```python
# convert to numeric values can't work with object values
numeric_vars = ['Global_active_power', 'Global_reactive_power', 'Voltage', 
           'Global_intensity', 'Sub_metering_1', 'Sub_metering_2', 'Sub_metering_3']
for i in numeric_vars:
    data[i] = pd.to_numeric(data[i])
```

```python
# replace -1 with mean of the column
imp = SimpleImputer(missing_values=-1, strategy='mean')
data[numeric_vars] = imp.fit_transform(data[numeric_vars])
```

```python
data.dtypes
```

```python
# check for Duplications
data["Duplications"]= data.duplicated()
data["Duplications"].value_counts()# false no duplications found
```

```python
data.drop(['Duplications'], axis=1, inplace=True)
```

```python
# Target value
temp1= (data['Global_active_power']*1000/60)
temp2= data['Sub_metering_1'] + data['Sub_metering_2'] + data['Sub_metering_3']
data['power_consumption']=temp1-temp2
data.head()
```

```python
# drop the unnecessary columns
data.drop(['Date', 'Time'], axis=1,inplace=True)
```

```python
# Distribution of the target variables
sns.histplot(data=data, x='power_consumption', bins=15, kde=True)
plt.show()
```

```python
data.dtypes
```

```python
column_names_array = data.columns.to_numpy()
column_names_array
```

```python
import math
n_cols=4
n_rows = math.ceil(len(column_names_array)/n_cols)
fig, ax = plt.subplots(n_rows, n_cols, figsize=(20, 8))
ax = ax.flatten()
for i, column in enumerate(column_names_array):
    sns.kdeplot(
        data[column],
        ax=ax[i], color='red'
)
```

```python
# correlation matrix 
corr = np.corrcoef(data.corr())
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(data.corr(), annot=True, mask=mask)
plt.show()
```
